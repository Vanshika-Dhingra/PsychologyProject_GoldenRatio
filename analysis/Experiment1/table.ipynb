```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Experiment1Data.csv')
df
```

```python
df.to_numpy().shape
```

```python
array = {
    'rectangle' : np.zeros(6),
    'parallelogram':np.zeros(6),
    'ellipse':np.zeros(6),
    'trapezium':np.zeros(6),
    'square':np.zeros(6),
    'circle':np.zeros(6),
}
```

```python
shapes = [
    'rectangle',
    'parallelogram',
    'ellipse',
    'trapezium',
    'square',
    'circle']
```

```python
arr = df.to_numpy()
arr
```

```python
for idx,col_name in enumerate(list(df.columns)):
    for shape in shapes:
#         print(col_name)
        if shape in col_name:
            new_arr = arr[:,idx]
            for i in new_arr:
                print(i)
                num = int(i[-1])-1
                array[shape][num] += 1
```

```python
ratios = [1,1.268,1.618,2.098,2.791,3.885]
```

```python
array
```

```python
for i,j in array.items():
    plt.plot(j)
```

```python
for i,j in array.items():
    array[i] /= array[i].sum()
```

```python
array
```

```python
df = pd.DataFrame(array)
df.index = ratios
df.columns =  [
    'Rectangle',
    'Parallelogram',
    'Ellipse',
    'Trapezium',
    'Square',
    'Circle']
df
```
